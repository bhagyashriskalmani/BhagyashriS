```python

"""AGRICULTURAL DATA ANALYSIS"""

def load_data():
    """Reads data from file and returns a list of dictionaries"""
    crops = []
    with open("crop.txt", "r") as file:
        for line in file:
            line = line.strip()
            if line:
                data = line.split(",")
                crop = {
                    "crop_name": data[0],
                    "season": data[1],
                    "area": float(data[2]),
                    "yield": float(data[3]),
                    "cost": float(data[4]),
                    "revenue": float(data[5])
                }
                crops.append(crop)
    return crops


def Kharif(crops):
    Kharif = []
    for crop in crops:
        if crop["season"] == "Kharif":
            Kharif.append(crop["crop_name"])
    return Kharif


def Rabi_crops(crops):
    Rabi = []
    for crop in crops:
        if crop["season"] == "Rabi":
            Rabi.append(crop["crop_name"])
    return Rabi
def calculated_profit(crops):
  for crop in crops:
   profit=crop["revenue"]-crop["cost"]

   print(crop["crop_name"],"--",round(profit,2))


def statistical_summary(crops):

   total_area=sum(crop["area"] for crop in crops)
   total_yield=sum(crop["yield"] for crop in crops)
   average_yield=total_yield/len(crops)
   print("\nstatistical summary")
   print("total area:",total_area)
   print("average yield:",round(average_yield,2))


def display_crops(crops):
    for crop in crops:
        for key, value in crop.items():
            print(f"{key}: {value}")
        print("----")


def write_report(filename, crops):
    with open(filename, "w") as file:
        file.write("AGRICULTURAL DATA ANALYSIS REPORT\n\n")
        for crop in crops:
            for key, value in crop.items():
                file.write(f"{key}: {value}\n")
            file.write("----\n")


def main():
    output_file = "analysis-report.txt"

    crops = load_data()
    print(crops)


    display_crops(crops)

    print("Kharif:", Kharif(crops))
    print("Rabi:", Rabi_crops(crops))
    print("\n\nprofit of crops")
    calculated_profit(crops)
    statistical_summary(crops)


    write_report(output_file, crops)
    print("\nAnalysis complete. Report written to", output_file)


main()

```
